## 3) Il y a des antennes pour lesquelles on a des factures mensuelles tandis que pour d'autres, on les reçoit moins régulièrement (trimestre, semestre, année). Ajoute une colonne à ton tableau qui calcule le nombre de jours entre le début de la période de facturation et la fin de la période de facturation.

```python
import datetime
df[['BEGIN','END']] = df[['BEGIN','END']].apply(pd.to_datetime)
df
```

```python
df['DUREE']=(df['END']-df['BEGIN']).dt.days.astype('Int64')
df
```

## 4) A partir de cette nouvelle colonne, crée une autre colonne qui nous donne la consommation journalière moyenne de la période.

```python
df.dtypes
```

### Espace dans le string, donc suppression ci-dessous

```python
df['KWH'] = df['KWH_PERIOD'].str[0:1] + df['KWH_PERIOD'].str[2:]
df
```

```python
df.dtypes
```

```python
df['KWH'] = pd.to_numeric(df['KWH'], errors='coerce').fillna(0).astype(np.int64)
df.dtypes
```

```python
df
```

```python
df['CONSOMMATION'] = df['KWH'] / df['DUREE']
df
```

## 6) Et information bonus si tu y arrives : les équipes aimeraient beaucoup avoir une liste de toutes les antennes avec, pour chaque antenne, le coefficient de variation de la consommation. Cela permettra d'avoir une idée, pour chaque antenne, de la dispersion de sa consommation d'énergie.

```python
df_cv = df[['SITE_NAME','PERIOD_EUR']]
df_cv['PERIOD_EUR'] = df_cv['PERIOD_EUR'].str.replace(' ', '')
df_cv['PERIOD_EUR'] = pd.to_numeric(df_cv['PERIOD_EUR'].str.replace(',', '.'))

df_cv_mean = df_cv.groupby(['SITE_NAME'])['PERIOD_EUR'].mean().to_frame(name='MEAN')
df_cv_mean.reset_index(inplace = True)

df_cv_std = df_cv.groupby(['SITE_NAME'])['PERIOD_EUR'].std().to_frame(name='STD')
df_cv_std.reset_index(inplace = True)

df_cv_final = pd.concat([df_cv_mean,df_cv_std['STD']], axis=1)

df_cv_final['CV'] = (df_cv_std['STD'] / df_cv_mean['MEAN'] * 100).round(2).astype(str) + '%'

df_cv_final = df_cv_final[['SITE_NAME','CV']]
print(df_cv_final)
df_cv_final.to_csv('coefvar.csv',index=False)
```
